Import libraries.


```python
from google.colab import files
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
```

Upload image.

```python
uploaded = files.upload()

img_path = next(iter(uploaded))
```

Plot saved image.

```python
img = image.load_img(img_path)

plt.imshow(img)
plt.axis('off')
plt.show()
```

Transform image to grayscale using the luminance formula: 0.299 × R + 0.587 × G + 0.114 × B and plot the result.

```python
width, height = img.size


gs_img = img.copy()


for x in range(width):
    for y in range(height):
        r, g, b = img.getpixel((x, y))
        gray = int(0.299 * r + 0.587 * g + 0.114 * b)
        gs_img.putpixel((x, y), (gray, gray, gray))


plt.imshow(gs_img)
plt.axis('off')
plt.show()
```

Transform the image to black and white by classifying pixel values based on a threshold, and plot the result.

```python
bw_img = img.copy()


for x in range(width):
    for y in range(height):
        gs = img.getpixel((x, y))[0]
        bw = 0 if gs < 128 else 255
        bw_img.putpixel((x, y), (bw, bw, bw))


plt.imshow(bw_img)
plt.axis('off')
plt.show()
```
